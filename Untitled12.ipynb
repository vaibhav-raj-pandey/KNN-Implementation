```python
def dist(XT,Xt):
    return np.sqrt(sum((XT-Xt)**2))
        
    
```

```python
def knn(XT,YT,XTest,k=5):
    print(XTest)
    vals=[]
    m=XT.shape[0]
    
    for i in range(m):
        d=dist(XT[i],XTest)
        vals.append((d,YT[i]))
        
    vals=sorted(vals)
    vals=vals[:k]
    
    vals=np.array(vals)
    new_vals= np.unique(vals[:,1],return_counts='true')
    
    index= new_vals[1].argmax()
    pred= new_vals[0][index]
    
    return pred

k=knn(dtX,dtY,tX[4])

print(k)   
    
    
    
```
